```python
import time
import numpy as np

size = 1_000_000

# Python list
a = list(range(size))
b = list(range(size))

start = time.time()
c = [a[i] + b[i] for i in range(size)]
end = time.time()

print("List addition time:", end-start)

# NumPy
a_np = np.arange(size)
b_np = np.arange(size)

start = time.time()
c_np = a_np + b_np
end = time.time()

print("NumPy addition time:",end-start)
```

```python
# Python list
start = time.time()
c = [a[i] * b[i] for i in range(size)]
end = time.time()
print("List multiplication time:", end-start)

# NumPy
start = time.time()
c_np = a_np * b_np
end = time.time()
print("NumPy multiplication time:",end-start)
```

```python
# Python
start = time.time()
dot_list = sum(a[i] * b[i] for i in range(size))
end = time.time()
print("List dot time:", end-start)

# NumPy
start = time.time()
dot_np = np.dot(a_np, b_np)
end = time.time()
print("NumPy dot time:",end-start)
```

```python
# Python lists
import random
M1 = [[random.random() for _ in range(1000)] for _ in range(1000)]
M2 = [[random.random() for _ in range(1000)] for _ in range(1000)]

start = time.time()
result = [[sum(a*b for a,b in zip(r,c)) for c in zip(*M2)] for r in M1]
end = time.time()
print("List matrix multiplication time:",end-start)
```

```python
# NumPy
M1_np = np.random.random((1000,1000))
M2_np = np.random.random((1000,1000))

start = time.time()
result_np = np.dot(M1_np, M2_np)
end = time.time()
print("NumPy matrix multiplication time:",end-start)
```
